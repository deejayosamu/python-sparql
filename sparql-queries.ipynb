```python
from SPARQLWrapper import SPARQLWrapper, JSON

# Set the SPARQL endpoint URL
sparql_endpoint_society = "http://localhost:3030/society/sparql"

# Create a SPARQLWrapper instance
sparql_society = SPARQLWrapper(sparql_endpoint_society)

# Set the SPARQL query
sparql_society.setQuery("""
    SELECT ?subject ?predicate ?object
    WHERE {
      ?subject ?predicate ?object
    }
    LIMIT 25
""")

# Specify the format of the results (JSON in this case)
sparql_society.setReturnFormat(JSON)

# Execute the query and get the results
results = sparql_society.query().convert()

# Print the results
for result in results["results"]["bindings"]:
    subject = result["subject"]["value"]
    predicate = result["predicate"]["value"]
    obj = result["object"]["value"]
    print(f"Subject: {subject}, Predicate: {predicate}, Object: {obj}")
```

```python
# Question : Let me know all living languages.

# Set the Query 
sparql_society.setQuery("""
    PREFIX kko: <http://kbpedia.org/kko/rc/>
    PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
    
    SELECT ?languages
    WHERE {
      ?languages rdfs:subClassOf kko:LivingLanguage.
}
""")

# Specify the format of the results (JSON in this case)
sparql_society.setReturnFormat(JSON)

# Execute the query and get the results
results = sparql_society.query().convert()

# Print the results
for result in results["results"]["bindings"]:
    language = result["languages"]["value"]
    print(language[26:])
```
